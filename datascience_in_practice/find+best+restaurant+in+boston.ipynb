```python
search_results = yelp_api.search_query(term='restaurant', location='boston')
```

```python
total = search_results['total']
```

```python
highest_offset = int(ceil(total/50))
```

```python
search_results['region']
```

```python
all_business = []
for i in range (1, highest_offset+1):
    print(i)
    try:
        search_results = yelp_api.search_query(term='restaurant', location='boston', offset = i)
        all_business.extend(search_results['businesses'])
    except:
        print('problem',i)
```

```python
locations = []
for biz in all_business[:500]:
    locations.append((biz['coordinates']['latitude'],biz['coordinates']['longitude']))
```

```python
import os
```

```python
from folium.plugins import MarkerCluster

m = folium.Map(location=locations,tiles='Cartodb Positron', zoom_start=1)

m.add_child(MarkerCluster(locations=locations))

m.save('1000_MarkerCluster.html')

m
```
