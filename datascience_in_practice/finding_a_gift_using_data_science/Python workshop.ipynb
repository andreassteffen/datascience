#history
* started late 1980s based on [ABC](http://homepages.cwi.nl/~steven/abc/) by Guido van Rossum @ __Centrum Wiskunde & Informatica__
    * nesting by indentation
    * addition of exception handling
    * module system
* code made public in 1991
* python 1.0 released in 1994
    * inclusion of functional programming commands *lambda, map, filter and reduce*
* python 2.0 released in 2000
    * full garbage collector
    * list comprehensions (advanced section)
    * unicode support (but problematic)

* Python 2.6 released in 2008 (6 years ago!!!!)
    * main python version of scientific linux distribution
* python 3.0 released in 2008
    * complete redesign to remove fundamental flaws
        * *reduce feature duplication by removing old ways of doing things*
    * no backwards compatibility
    * print becomes a built-in function not a statement
        * print "hello world!" vs print("hello world!")
    * unicode!
* python 2.7 released in 2010
    * the python version till 2020, which will coexist with python 3.0

suggested reading [python history blog](http://python-history.blogspot.co.at)

# Practical coding tips
* some useful tips [python style guide](http://legacy.python.org/dev/peps/pep-0008/#programming-recommendations)
* coding
    * interactive computing in ipython or ipython notebook
    * for programs develop in sublime text or pycharm
    * set tab to 4 spaces
    * mostly __print__ debugging, otherwise [pdb](https://docs.python.org/2/library/pdb.html)
* spelling
    * no camel case, better: get_gene_expression_data()
    * variables spelled small
    * classes capital letter
    * globals large letters
* modules
    * never from xyz import \*, better: *import pandas as pd* __or__ *from os import getcwd*

# Container

```python
# List
a = [1,2,3,4,'test', [1,2,3,4], {'test':1, 'test2': 3}] # you can mix types, have containers in containers
# indexing and slicing
print a[6],a[1:4]

# get index
print a.index(1)

# change element
a[6]='no more dict'
print a

# sorting
a = [3,4,1,2]
print sorted(a)
print a
a.sort()
print a

# operators
print 10* a
print a+['TEST','test','TeSt']
```

```python
# String

ausspruch = 'Der Kandidat selbst hat glaubhaft erklaert, dass er mit seiner Darbietung keinerlei Bezug zum Horst-Wessel-Lied herstellen wollte und ihm die Aehnlichkeit auch nicht bewusst war.'
print ausspruch

# reverse
print ausspruch[::-1]

# slice
print ausspruch[4:12]
print ausspruch[::2]

# get index
print ausspruch.index('Kandidat') # first hit
print ausspruch.find('er') # first from left
print ausspruch.rfind('er') # first from right

# replace
print ausspruch.replace('keinerlei','deutlichen').replace('auch nicht','sehr')


# operators
print "ASDF"+"asdf"
print 2*"ASDF"

# formatting
print "%s\t%s" % ('Andreas', 'Steffen')
data = {"first_name":'Andreas', "last_name":'Steffen', 'employment': 4.782934}
s = "first_name: %(first_name)s last_name: %(last_name)s, employed since %(employment)0.2f years" % data
print s
```

```python
# dictionary
# --> non-ordered hashes

# initialize
my_dict = {} 
my_dict = dict()
my_dict = {'test0': 0, 'test1': 1}

# assign values
my_dict['test2'] = 2
print my_dict
my_dict['test3'] = 3

# keys
print my_dict.keys()

# values
print my_dict.values()

# items
print my_dict.items()

# set default value, e.g. list
my_dict.setdefault(1,[]).append('oh oh oh!')
my_dict.setdefault(1,[]).append('oh oh oh!')
print my_dict

# check for key
print 'test1' in my_dict, 'test42' in my_dict

# advanced
keys = ['Andreas', 'Angela', 'Bill']
values = ['Steffen', 'Merkel', 'Evans']
my_dict = dict(zip(keys, values))
print my_dict
```

```python
# set

# initialize
a = set()
print a

# add
a.add(1)

# add multiple
a.update([1,2,3,4,5,6])
print a

# check whether in
print 1 in a

# intersection
b = set([4,5,6,7,8])
print a.intersection(b)

# difference
print a.difference(b)

# union
print a.union(b)
```
